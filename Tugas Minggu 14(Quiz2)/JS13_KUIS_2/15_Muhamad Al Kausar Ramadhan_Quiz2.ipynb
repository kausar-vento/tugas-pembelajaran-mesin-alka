```python
#memanggil library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#df digunakan sebagai variable untuk memanggil file excel yang bernama dataset_quiz yang ada di folder data
df = pd.read_excel('data/dataset_quiz.xlsx')
df.head(100)
```

<b>Random Search</b>

```python
#Memanggil RepeatedKFold
from sklearn.model_selection import RepeatedKFold
# Inisiasi obyek RepeatedKFold dengan split 4, repeats 3, dan random_state = 42
rkf = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

```python
#Memanggil RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV

# Inisiasi obyek DecisionTreeClassifier
dtc = DecisionTreeClassifier()

#set parameter dengan criterion : gini, entropy, log_loss (wajib) dengan max depth di antara 5 - 11
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11))
}

# Inisiasi obyek RandomizedSearchCV dengan estimator yang dipakai adalah dtc
# value parameter nya adalah params, cv adalah rkf sebagai validasi silang K-fold
random = RandomizedSearchCV(dtc, param_distributions=params, cv=rkf)
%timeit random.fit(X, y)

#mengambil nilai evaluasi dari X dan y
score = random.score(X,y)

#cetak hasilnya
print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {random.best_params_}')
```

<b>Grid Search</b>

```python
# Memanggil GridSearchCV dan SVC
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Inisiasi objek SVC
svc = SVC()

#mendefinisikan range parameter
params = {
    'kernel': ['rbf', 'linear', 'poly'],
    'C': [3],
    'degree': [1]
}

# Inisiasi obyek GridSearchCV estimator yang dipakai adalah SVC, 
# value parameter nya adalah params, cv adalah rkf sebagai validasi silang K-fold
grid = GridSearchCV(svc, param_grid=params, cv=rkf, n_jobs=-1)

#proses pengambilan nilai scorenya
%timeit grid.fit(X, y)
score = grid.score(X,y)

#mendapatkan hasilnya
print(f'Hasil evaluasi Grid Search: {score}')
print(f'Konfigurasi hyperparameter Grid Search: {grid.best_params_}')
```
