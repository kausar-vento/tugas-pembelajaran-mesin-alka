```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('dataset/mushrooms.csv')
data.head()
```

```python
lbE = LabelEncoder()

for column in data.columns:
    data[column] = lbE.fit_transform(data[column])

data.head()
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
dt_entropy = DecisionTreeClassifier(criterion='entropy')
dt_gini = DecisionTreeClassifier(criterion='gini')

dt_entropy.fit(X_train, y_train)
y_pred_entropy = dt_entropy.predict(X_test)

dt_gini.fit(X_train, y_train)
y_pred_gini = dt_gini.predict(X_test)

acc_entropy = accuracy_score(y_test, y_pred_entropy)
acc_gini = accuracy_score(y_test, y_pred_gini)

print(f'Akurasi Entropi: {acc_entropy}')
print(f'Akurasi Gini: {acc_gini}')
```

```python
#Logistic Regression Classification
lr = LogisticRegression(solver="liblinear")
lr.fit(X_train, y_train)
hasilLogistic = lr.score(X_test, y_test)

print(f'Hasil Logistic Regresion: {hasilLogistic}')
```

```python
nb = GaussianNB()
nb.fit(X_train,y_train)
hasilNb = nb.score(X_test, y_test)
print(f'Hasil Naive Bayes: {hasilNb}')
```
