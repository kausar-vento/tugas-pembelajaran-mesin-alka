<h2>JAWABAN NOMOR 1</h2>

```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/wbc.csv')
df = df.iloc[:,:-1]

df.head()
```

```python
X = df.iloc[:, 2:]
y = df['diagnosis']

X.head()
```

```python
from sklearn.model_selection import RepeatedKFold
rkf = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

dtc = DecisionTreeClassifier()

params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11))
}

random = RandomizedSearchCV(dtc, param_distributions=params, cv=rkf)
%timeit random.fit(X, y)

score = random.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {random.best_params_}')
```

<h2>JAWABAN NOMOR 2</h2>

<h3>RANDOM SEARCH</h3>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC

svc = SVC()
params = {
    'kernel': ['rbf', 'linear', 'poly'],
    'C': [3],
    'degree': [1]
}

random = RandomizedSearchCV(svc, param_distributions=params, cv=rkf, n_jobs=-1, )
%timeit random.fit(X, y)

score = random.score(X,y)

print(f'Hasil evaluasi Random Search: {score}')
print(f'Konfigurasi hyperparameter Random Search: {random.best_params_}')
```

<h3>GRID SEARCH</h3>

```python
from sklearn.model_selection import GridSearchCV
params = {
    'kernel': ['rbf', 'linear', 'poly'],
    'C': [3],
    'degree': [1]
}

grid = GridSearchCV(svc, param_grid=params, cv=rkf, n_jobs=-1)

%timeit grid.fit(X, y)
score = grid.score(X,y)

print(f'Hasil evaluasi Grid Search: {score}')
print(f'Konfigurasi hyperparameter Grid Search: {grid.best_params_}')
```
