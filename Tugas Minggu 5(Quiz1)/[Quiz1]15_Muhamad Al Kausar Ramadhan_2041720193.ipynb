```python
# the independent variables set
data_set = new_df[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]
  
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = data_set.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(data_set.values, i)
                          for i in range(len(data_set.columns))]
  
print(vif_data)
```
